## requests package

# Making a GET request

```python
import requests

def print_response(resp):
    if(resp != None):
        print("Status Code: {}".format(resp.status_code))
    #if(resp.status_code != None):
        #print("Reason for status: {}".format(resp.raise_for_status()))
        print("Reason: {}".format(resp.reason))
        print("Encoding: {}".format(resp.encoding))
        print("Content Length: {}".format(len(resp.text)))
        print(resp.text)
    else:
        print("Response object is NULL")

resp = requests.get("http://www.dynamicdeploy.com")
print_response(resp)
```

# Make a POST request

```python
data = {"email":"info@tutsplus.com","password":"12345"}
resp = requests.post("http://www.dynamicdeploy.com", params = data)
print_response(resp)
```

## Make a PUT request

```python
data= {"name":"tutsplus", "telephone":"12345"}
resp = requests.put("http://www.dynamicdeploy.com", params= data)
print_response(resp)
```

## Make a DELETE Request

```python
data= {'name':'Tutsplus'}
url = "https://www.contact.com/api/"
resp = requests.delete(url, params= data)
print_response(resp)
```

### Retrieve internet resource and store it

```python
import urllib.request
filename, headers = urllib.request.urlretrieve('http://python.org/')
html = open(filename)
print(html)
```

```python
# imported the urllib library
import urllib
from urllib.parse import urlparse
import os
def getfilenamefromurl(url):
    o = urlparse(url)
    path = o.path
    fpath, filename = os.path.split(path)
    return fpath, filename

def download_image(url, filename=None):
    if filename == None:
        fpath,filename=getfilenamefromurl(url)
  
    # Copy a network object to a local file
    urllib.request.urlretrieve(url, filename)
    print("{} saved".format(filename))
    
url = 'https://www.python.org/static/opengraph-icon-200x200.png'
 
download_image(url)

# downloading with requests
 
# import the requests library
import requests
 
# download the url contents in binary format
r = requests.get(url)
fpath,filename=getfilenamefromurl(url)
# open method to open a file on your system and write the contents
with open(filename, "wb") as code:
    code.write(r.content)
    print("{} saved".format(filename))
```

### Download PDF Files With Python

```python
url = 'https://static.googleusercontent.com/media/www.google.com/en//googleblogs/pdfs/google_predicting_the_present.pdf'
 

# import the urllib package
import urllib
 
# Copy a network object to a local file
download_image(url, "tutorial.pdf")
 
 # import the requests library
import requests
 
# download the file contents in binary format
r = requests.get(url)
fpath,filename=getfilenamefromurl(url)
# open method to open a file on your system and write the contents
with open(filename, "wb") as code:
    code.write(r.content)
    print("{} saved".format(filename))
```
