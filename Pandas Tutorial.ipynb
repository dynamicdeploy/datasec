```python
import pandas as pd

reviews = pd.read_csv("ign.csv")
#returns 5 records by default
print(reviews.head())
print("-----------------------------")
# returns number of rows and columns
print(reviews.shape)
print("-----------------------------")
print(reviews.tail())
print("-----------------------------")
print(reviews.info())
print("-----------------------------")
```

## Indexing using numbers

### The iloc method allows us to retrieve rows and columns by position.

```python
#return rows from 0-5. First row is 0
print(reviews.iloc[0:5,:])
```

```python
#return 5 rows and 1 column
print(reviews.iloc[0:5,0:1])
```

```python
# return the first 5 rows, and all of the columns for those rows.
print(reviews.iloc[:5,:])
```

```python
#the entire DataFrame.
print(reviews.iloc[:,:])
```

```python
#rows from position 5 onwards, and columns from position 5 onwards.
print(reviews.iloc[5:,5:])
```

```python
#the first column, and all of the rows for the column.
print(reviews.iloc[:,0])
print(reviews.iloc[:,:1])
```

```python
#rows from position 5 onwards, and columns from position 5 onwards.
print(reviews.iloc[5:,5:])
```

```python
#the first column, and all of the rows for the column.
print(reviews.iloc[:,:1])
print(reviews.iloc[:,0])
```

```python
# the 10th row, and all of the columns for that row.
print(reviews.iloc[9,:])
```

```python
#remove first column
reviews = reviews.iloc[:,1:]
reviews.head()
```

## Indexing using labels

### The loc method allows us to retrieve rows and columns by labels.

```python
reviews.loc[0:5,:]
```

```python
# the difference between iloc and loc is that loc returns row labels instead of index. compare with example
reviews.iloc[0:5,:]
```

```python
reviews.index
```

```python
#Get row 10 to row 20 of reviews, and assign the result to some_reviews.
some_reviews = reviews.iloc[10:20,:]
some_reviews.head()
```

```python
#specify column labels in the loc method to retrieve columns by label instead of by position.
reviews.loc[:5,"score"]
```

```python
#more than one column at a time by passing in a list
reviews.loc[:5,["score", "release_year"]]
```

```python
#retrieve a whole column by label
reviews["score"]
```

```python
#lists of columns with this method
reviews[["score", "release_year"]]
```

```python
reviews[["score", "release_year"]]
```

## Series  = one column

```python
#A single column representing a Pandas Series object. 
#A DataFrame stores tabular data, but a Series stores a single column or row of data.
print(type(reviews["score"]))
print(type(reviews[["score", "release_year"]]))
```

```python
#To create a Series, we pass a list or NumPy array into the Series object when we instantiate it:
s1 = pd.Series([1,2])
s1
```

```python
s2 = pd.Series(["Boris Yeltsin", "Mikhail Gorbachev"])
s2
```

## Creating A DataFrame in Pandas

```python
#create pandas from 2 series
pd.DataFrame([s1,s2])
```

```python
#different way with same result
pd.DataFrame(
    [
        [1,2],
        ["Boris Yeltsin", "Mikhail Gorbachev"]
    ]
)
```

```python
#specify column labels
pd.DataFrame(
    [
        [1,2],
        ["Boris Yeltsin", "Mikhail Gorbachev"]
    ],
    columns=["column1", "column2"]
)
```

```python
#specify row labels
frame = pd.DataFrame(
    [
        [1,2],
        ["Boris Yeltsin", "Mikhail Gorbachev"]
    ],
    index=["row1", "row2"],
    columns=["column1", "column2"]
)
frame
```

```python
# index the dataframe using labels
frame.loc["row1":"row2", "column1"]
```

```python
#passing columns as a dictionary
frame = pd.DataFrame(
    {
        "column1": [1, "Boris Yeltsin"],
        "column2": [2, "Mikhail Gorbachev"]
    }
)
frame
```

```python
#same methods apply on a Series object that we can on a DataFrame
reviews["title"].head()
```

### View unique values in a series

```python
reviews["platform"].value_counts(dropna=False)
```

```python
#series mean
reviews["score"].mean()
```

```python
#dataframe mean
reviews.mean()
```

```python
# use axis keyword argument in order to compute the mean of each row or of each column
# axis=1 to compute the mean of each row
reviews.mean(axis=1)
```

```python
reviews.iloc[0,:]
```

```python
#pandas.DataFrame.corr  finds the correlation between columns in a DataFrame.
# none of our numeric columns correlates with score, meaning that release timing doesn't linearly relate to review score.
print("Correlation:{0}".format(reviews.corr()))
print("Count:{0}".format(reviews.count()))
print("Max:{0}".format(reviews.max()))
print("Min:{0}".format(reviews.min()))
print("Mean:{0}".format(reviews.mean()))
print("Std:{0}".format(reviews.std()))
print("Median:{0}".format(reviews.median()))
print("All Stats")
print(reviews.describe())
```

```python
#Stats on a series
print("Score stats")
print("Count:{0}".format(reviews["score"].count()))
print("Max:{0}".format(reviews["score"].max()))
print("Min:{0}".format(reviews["score"].min()))
print("Mean:{0}".format(reviews["score"].mean()))
print("Std:{0}".format(reviews["score"].std()))
print("Median:{0}".format(reviews["score"].median()))
```

## Boolean Indexing in Pandas

```python
score_filter = reviews["score"] > 7
score_filter
```

```python
filtered_reviews = reviews[reviews["score"] > 7]
filtered_reviews.head()
```

```python
#using mulitple filters
xbox_one_filter = (reviews["score"] > 7) & (reviews["platform"] == "Xbox One")
filtered_reviews = reviews[xbox_one_filter]
filtered_reviews
len(filtered_reviews)
```

## Pandas Plotting

Uses pandas.DataFrame.plot --> Matplotlib

### Score histogram for each console

```python
# plot method defaults to drawing a line graph.
# pass in the keyword argument kind="hist" to draw a histogram instead.
# %matplotlib inline to set up plotting inside a Jupyter notebook
%matplotlib inline
#create filter
xbox_one_filter = reviews["platform"] == "Xbox One"
#apply filter
filtered_reviews = reviews[xbox_one_filter]
#plot the score histogram
filtered_reviews["score"].plot(kind='hist')
```

```python
#all in one line of code
reviews[reviews["platform"] == "PlayStation 4"]["score"].plot(kind="hist")
```

```python
#can also use hist()  function directly
filtered_reviews["score"].hist()
```
